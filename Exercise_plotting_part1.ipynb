### Exercise 2

This dataset actually contains 13 separate example datasets, each with two variables named `example[number]_x` and `example[number]_y`. 

In order to get a better sense of what these datasets look like, write a loop that iterates over each example dataset (numbered 1 to 13) and print out the mean and standard deviation for `example[number]_x` and `example[number]_y` for each dataset. 

For example, the first iteration of this loop might return something like:

```
Example Dataset 1: 
Mean x: 23.12321978429576, 
Mean y: 98.23980921730972, 
Std Dev x: 21.2389710287, 
Std Dev y: 32.2389081209832,
Correlation: 0.73892819281
```

(Though you shouldn't get those specific values)

````python
# Looping over each example dataset (from 1 to 13)
for i in range(1, 14):
    # Extracting the relevant columns
    x_col = f"example{i}_x"
    y_col = f"example{i}_y"

    # Computing the mean, standard deviation, and correlation
    mean_x = datasaurus[x_col].mean()
    mean_y = datasaurus[y_col].mean()
    std_x = datasaurus[x_col].std()
    std_y = datasaurus[y_col].std()
    correlation = datasaurus[x_col].corr(datasaurus[y_col])

    # Printing the results
    print(f"Example Dataset {i}: ")
    print(f"Mean x: {mean_x},")
    print(f"Mean y: {mean_y},")
    print(f"Std Dev x: {std_x},")
    print(f"Std Dev y: {std_y},")
    print(f"Correlation: {correlation}\n")
````

````python
# Display the results in a table format
# Initializing a list to store the results
results = []

# Looping over each example dataset (from 1 to 13)
for i in range(1, 14):
    # Extracting the relevant columns
    x_col = f"example{i}_x"
    y_col = f"example{i}_y"

    # Computing the mean, standard deviation, and correlation
    mean_x = datasaurus[x_col].mean()
    mean_y = datasaurus[y_col].mean()
    std_x = datasaurus[x_col].std()
    std_y = datasaurus[y_col].std()
    correlation = datasaurus[x_col].corr(datasaurus[y_col])

    # Appending the results to the list
    results.append(
        {
            "Dataset": f"Example {i}",
            "Mean x": mean_x,
            "Mean y": mean_y,
            "Std Dev x": std_x,
            "Std Dev y": std_y,
            "Correlation": correlation,
        }
    )

# Converting the results list to a DataFrame
results_df = pd.DataFrame(results)

# Displaying the DataFrame
results_df
````

### Execise 4

Write a loop that iterates over these example datasets, and using Altair library, plot a simple scatter plot of each dataset with the `x` variable on the x-axis and the `y` variable on the y-axis.

Hint: When writing this type of code, it is often best to start by writing code to do what you want for the first iteration of the loop. Once you have code that works for the first example dataset, then write the full loop around it. 

Hint 2: To force Jupyter to display your charts when they're generated within a loop, use the method `.show()` (e.g. `my_chart.show()`).

Hint 3: You will need to change the range of the axes to make the plots look good!

````python
import altair as alt
from IPython.display import display

# Looping over each example dataset (from 1 to 13)
for i in range(1, 14):
    # Extracting the relevant columns
    x_col = f"example{i}_x"
    y_col = f"example{i}_y"

    # Creating a new dataframe with the subset of the data
    subset_df = datasaurus[[x_col, y_col]].rename(columns={x_col: "x", y_col: "y"})

    # Creating the scatter plot
    chart = (
        alt.Chart(subset_df)
        .mark_point()
        .encode(
            x=alt.X("x:Q", scale=alt.Scale(domain=(0, 100))),
            y=alt.Y("y:Q", scale=alt.Scale(domain=(0, 100))),
        )
        .properties(title=f"Example Dataset {i}")
    )

    # Displaying the plot
    display(chart)
````

### Exercise 6

Load the World Development Indicator data used in the [plotting reading](https://nickeubank.github.io/practicaldatascience_book/notebooks/class_5/week_2/20_plotting_with_seaborn.html). Rather than picking a single year, pick a single country and look at how GDP per capita and one of the other variables in that dataset have evolved together over time. 

Make any adjustments to the functional forms of your variables and/or axes needed to make the figure legible. 

````python
wdi_data = (
    "https://raw.githubusercontent.com/nickeubank/"
    "practicaldatascience/master/Example_Data/wdi_plotting.csv"
)
world = pd.read_csv(wdi_data)
world.sample(5)
````

````python
import seaborn.objects as so

# Choose a single country China
country_name = "China"

country = world[world["Country Name"] == country_name]

# Create a plot that illustrate how GDP per capita and population in China have evolved together over time.
(
    so.Plot(
        country,
        x="GDP per capita (constant 2010 US$)",
        y="Population, total",
    )
    .add(so.Dot())
    .label(
        x="GDP Per Capita",
        y="Population",
        title="Evolution of GDP per Capita and Population in China",
    )
)
````

> + This graph illustrates how GDP per capita and population in China have evolved together over time.
> + From the graph, it shows that as the GDP per capita in China increased over time, the population also increased.

````python
import seaborn.objects as so

# Choose a single country China
country_name = "China"

country = world[world["Country Name"] == country_name]

# Create a plot that illustrate how GDP per capita and population in China have evolved together over time.
(
    so.Plot(
        country,
        x="Year",
        y="GDP per capita (constant 2010 US$)",
        pointsize="Population, total",
    )
    .add(so.Dot())
    .label(
        x="Year",
        y="GDP Per Capita",
        pointsize="Population",
        title="Evolution of GDP per Capita and Population in China (1971-2020)",
    )
)
````

> + This graph illustrates how GDP per capita and population in China have evolved together over time (year 1970 to 2020).
> + From the graph, it shows that as the GDP per capita in China increased from year 1970 to 2020, the population also increased.

### Exercise 7

Now add a second series. Facet your plot so that the two subplots are positioned so that they are effectively sharing the same time axes (e.g., if you draw a line up from 2010 on one plot, you get to 2010 on the other).

Rather than telling you exactly how to do it, however, I'll point you to the [seaborn tutorial](https://seaborn.pydata.org/tutorial/objects_interface#faceting-and-pairing-subplots). It has examples that don't do exactly what you want, but should be close enough you can guess-and-check to the solution you want!

Use your detective skills (and some guess and check work) to figure out how to get it to work!


````python
(
    so.Plot(country, x="Year")
    .pair(
        y=[
            "GDP per capita (constant 2010 US$)",
            "Population, total",
        ]
    )
    .add(so.Dots())
    .label(
        y0="GDP Per Capita",
        y1="Population",
        x="Year",
        title="Evolution in China(1971-2020)",
    )
)
````

> + The graph displays the trends in two variables, GDP per capita (constant 2010 US dollars) and population, for China from 1971 to 2020, allowing for a clear visualization of their behavior over time.
> + From the graph, it shows that as the GDP per capita in China increased from 1971 to 2020, the population also increased.
